```python
import json
internet=[]
ac=[]
club=[]
intercom=[]
pool=[]
cpa=[]
fs=[]
servant=[]
security=[]
sc=[]
gp=[]
park=[]
rwh=[]
stp=[]
hk=[]
pb=[]
vp=[]
for i in property_data['amenities']:
    obj=json.loads(i)
    internet.append(obj['INTERNET'])
    ac.append(obj['AC'])
    if 'CLUB' in obj.keys():
        club.append(obj['CLUB'])
    else:
        club.append(False)
    intercom.append(obj['INTERCOM'])
    if 'FS' in obj.keys():
        fs.append(obj['FS'])
    else:
        fs.append(False)
    if 'CPA' in obj.keys():
        cpa.append(obj['CPA'])
    else:
        cpa.append(False)
    if 'SERVANT' in obj.keys():
        servant.append(obj['SERVANT'])
    else:
        servant.append(False)
    security.append(obj['SECURITY'])
    sc.append(obj['SC'])
    if 'GP' in obj.keys():
        gp.append(obj['GP'])
    else:
        gp.append(False)
    park.append(obj['PARK'])
    if 'RWH' in obj.keys():
        rwh.append(obj['RWH'])
    else:
        rwh.append(False)
   # rwh.append(obj['RWH'])
    if 'STP' in obj.keys():
        stp.append(obj['STP'])
    else:
        stp.append(False)
  #  stp.append(obj['STP'])
    hk.append(obj['HK'])
    pb.append(obj['PB'])
    if 'VP' in obj.keys():
        vp.append(obj['VP'])
    else:
        vp.append(False)
  #  vp.append(obj['VP'])

```

```python
d1=pd.DataFrame({'ac':ac,'internet':internet,'intercom':intercom,'vp':vp,'stp':stp,'rwh':rwh,'gp':gp,'servant':servant,'cpa':cpa,'fs':fs,'club':club,'fs':fs,'security':security,'sc':sc,'park':park,'hk':hk,'pb':pb})
d1=d1.astype(int)
new_data=pd.concat([property_data,d1],axis=1)
new_data
```
